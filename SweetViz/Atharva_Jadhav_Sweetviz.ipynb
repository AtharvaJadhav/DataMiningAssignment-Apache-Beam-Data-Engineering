# Titanic Dataset - Automatic Exploratory Data Analysis

In this notebook, we'll be exploring the classic Titanic dataset using the Sweetviz library for automatic EDA. Our goal is to quickly understand the structure, relationships, and anomalies in the data.


```python
# Installation of Sweetviz
!pip install sweetviz
```

```python
# Importing necessary libraries
import sweetviz as sv
import pandas as pd
```

```python
# Load the Titanic dataset
titanic_data = pd.read_csv('train.csv')

# Display the first few rows of the dataset
titanic_data.head()
```

```python
# Analyzing the dataset
report = sv.analyze(titanic_data)

# Display the report
report.show_html('titanic_report.html')
```

The Sweetviz library provides a quick and comprehensive overview of the dataset's distribution, correlations, and other statistics. From the generated report, we can derive initial insights and decide on further steps for data preprocessing and modeling.
